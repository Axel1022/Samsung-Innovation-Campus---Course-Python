# Actividad: Creación de una Base de Datos en Python

## Descripción

A continuación debes crear una base de datos simple utilizando listas en Python. Una base de datos es una colección de datos organizados que se utilizan para almacenar y recuperar información. En este caso, simularemos una base de datos de estudiantes con información básica.

## Instrucciones

1. Crea un programa en Python que permita a los usuarios realizar las siguientes acciones:

   a. Agregar un nuevo estudiante a la base de datos. Cada estudiante debe tener al menos un nombre, un número de identificación y una edad. Puedes utilizar una lista para almacenar estos datos.

   b. Mostrar la lista de estudiantes en la base de datos.

   c. Buscar un estudiante por su número de identificación y mostrar su información.

   d. Actualizar la información de un estudiante existente en la base de datos.

   e. Eliminar un estudiante de la base de datos.

2. Utiliza listas de Python para almacenar la información de los estudiantes. Por ejemplo, puedes crear una lista de diccionarios, donde cada diccionario representa un estudiante con sus datos.

3. Ofrece un menú interactivo que permita a los usuarios elegir las acciones que desean realizar en la base de datos.

4. Implementa la lógica necesaria para cada acción, como agregar, buscar, actualizar o eliminar estudiantes.

## Ejemplo de Base de Datos

A continuación, se muestra un ejemplo de cómo podría verse una base de datos de estudiantes utilizando listas y diccionarios:

```python
base_de_datos = [
    {"nombre": "Juan", "id": 1, "edad": 20},
    {"nombre": "María", "id": 2, "edad": 22},
    {"nombre": "Carlos", "id": 3, "edad": 21}
]

Desarrolle el código en la siguiente lista.

````python
#Recuerde agregar comentarios donde se describa el funcionamiento de las diferentes secciones.
import msvcrt
import os

base_de_datos = [
    {"nombre": "Juan", "id":1, "edad": 20},
    {"nombre": "María", "id":2, "edad": 22},
    {"nombre": "Carlos", "id":3, "edad": 21}
]
# Para colocar el indice automatico
def siguiente_indice ():
    #Optengo el ultimo indice y le sumo 1
    ultimo_studiant= base_de_datos[-1]
    siguiente_indice= ultimo_studiant["id"] + 1
    return siguiente_indice

def limpiar_pantalla():
    os.system('cls')

def esperar_tecla():
    print("Presiona una tecla para continuar...")
    msvcrt.getch()

#Pide el id
def tell_me_id():
    print("Ingrese id del estudiante")
    id = int(input(">>> "))
    return id

#Verifica si el id del estudiante se encuentra en la base de datos
#Retorna el id o None
def search_id(id:int):
    for student in base_de_datos:
        if student["id"] == id:
            return student
    return None

#Pide los datos
def tell_me_data():
    print("Ingrese nombre del estudiante")
    nombre = input(">>> ")
    print("Ingrese edad del estudiante")
    edad = input(">>> ")
    return nombre, edad

#Crea el diccionario con los datos
def text_to_dict( nombre, edad, id):
    datos_dict = {"nombre": nombre,"id":id, "edad": edad}
    return datos_dict

#Agregar un nuevo estudiante a la base de datos.
def añadir_estudiante():
    limpiar_pantalla()
    print("""
=============================
     Añadir Estudiante
=============================
          """)
    nombre , edad =  tell_me_data()
    id = siguiente_indice()
    nuevo_estudiante = text_to_dict(nombre, edad, id)
    base_de_datos.append(nuevo_estudiante)
    limpiar_pantalla()
    print("""
=============================
     Añadir Estudiante
=============================
          """)
    print ("\nEstudiante añadido correctamente.")
    esperar_tecla()

#Mostrar la lista de estudiantes en la base de datos.
def view_studiant():
    limpiar_pantalla()
    print("""
=============================
    Lista de Estudiante
=============================
          """)
    for student in base_de_datos:
        print("==========================================================")
        print(f" \n{student}\n")
        print("==========================================================")
    esperar_tecla()

#Buscar un estudiante por su número de identificación y mostrar su información.
def search_student():
    limpiar_pantalla()
    print("""
=============================
     Buscar Estudiante
=============================
          """)
    id_ingresado = tell_me_id()
    estudiante_encontrado = search_id(id_ingresado)

    limpiar_pantalla()
    print("""
=============================
     Buscar Estudiante
=============================
          """)
    if estudiante_encontrado:
        print("Estudiante encontrado:", estudiante_encontrado)
    else:
        print("Estudiante no encontrado")
    esperar_tecla()

#Actualizar la información de un estudiante existente en la base de datos.
def update():
    limpiar_pantalla()
    print("""
=============================
    Actualizar Estudiante
=============================
          """)
    id = tell_me_id()
    encontrado = search_id(id)
    if encontrado:
        nombre, edad = tell_me_data()
        data_actualizada_dict = text_to_dict(nombre, edad, id)
        for estudiante in base_de_datos:
            if estudiante["id"]==id:
                estudiante.update(data_actualizada_dict)
                limpiar_pantalla()
                print("""
=============================
    Actualizar Estudiante
=============================
          """)
                print ("Estudiante actualizado correctamente.")
                esperar_tecla()
                break
    else:
        print("No encontrado")
        esperar_tecla()

# Eliminar un estudiante de la base de datos.
def delete():
    limpiar_pantalla()
    print("""
=============================
    Eliminar Estudiante
=============================
          """)
    id = tell_me_id()
    encontrado = search_id(id)
    if encontrado:
        for estudiante in base_de_datos:
            if estudiante["id"]==id:
                confirmacion = input("\n¿Estás seguro de eliminar a " + estudiante["nombre"] + "? (S/N): ").lower()
                if confirmacion == 's':
                    base_de_datos.remove(estudiante)
                    limpiar_pantalla()
                    print("""
=============================
    Eliminar Estudiante
=============================
          """)
                    print("\nEstudiante eliminado correctamente.")
                    break
                else:
                    print("\nOperación de eliminación cancelada.")
    else:
        print ("Estudiante no encontrado.")
    esperar_tecla()

while True:
    limpiar_pantalla()
    print('''
=============================
          Menú:
=============================
  0. Salir
  1. Agregar Estudiante
  2. Buscar Estudiante
  3. Actualizar Estudiante
  4. Eliminar Estudiante
  5. Listar Estudiantes
=============================
    ''')
    try:
        opcion = int(input(">>> "))
        if opcion == 0:
            break
        elif opcion == 1:
            añadir_estudiante()
        elif opcion == 2:
            search_student()
        elif opcion == 3:
            update()
        elif opcion == 4:
            delete()
        elif opcion == 5:
            view_studiant()
        else:
            print("\nOpción no válida. Intente de nuevo.")
            esperar_tecla()
    except ValueError:
        print("\nError: Ingrese un número válido.")
        esperar_tecla()
````
