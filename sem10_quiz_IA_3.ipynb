## Q 03-02

### ¿Cuál es una sintaxis correcta para verificar el número de dimensiones en una matriz?

1. np.dim()
2. np.ndim()
3. np.ndim
4. np.dim

**Respuesta:**

````python

matriz_1d = np.array([1, 2, 3, 4, 5])
matriz_2d = np.array([[1, 2, 3], [4, 5, 6]])

num_dim_1d = np.ndim(matriz_1d)
num_dim_2d = np.ndim(matriz_2d)

# Resultados
print("Número de dimensiones de la matriz 1D:", num_dim_1d)
print(matriz_1d)
print("Número de dimensiones de la matriz 2D:", num_dim_2d)
print(matriz_2d)

````

1. Agregue una nueva variable 𝐴𝑚𝑜𝑢𝑛𝑡=𝑈𝑛𝑖𝑡𝑃𝑟𝑖𝑐×𝑈𝑛𝑖𝑡𝑠Amount=UnitPric×Units.
2. Precio unitario medio de cada región. Utilice el método groupby().
3. Precio unitario medio de cada región. Usa el método pivot_table().

**Respuesta:**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


data = {
    'Region': ['A', 'B', 'A', 'C', 'B', 'C'],
    'UnitPrice': [10, 20, 15, 12, 18, 22],
    'Units': [100, 150, 200, 120, 180, 250]
}
df = pd.DataFrame(data)

# 1. Agregar una nueva variable llamada Amount
df['Amount'] = df['UnitPrice'] * df['Units']

# 2. Precio unitario medio de cada región utilizando groupby()
precio_unitario_medio_por_region = df.groupby('Region')['UnitPrice'].mean()
print("Precio unitario medio de cada región utilizando groupby():")
print(precio_unitario_medio_por_region)

# 3. Precio unitario medio de cada región utilizando pivot_table()
precio_unitario_medio_por_region_pivot = df.pivot_table(index='Region', values='UnitPrice', aggfunc='mean')
print("\nPrecio unitario medio de cada región utilizando pivot_table():")
print(precio_unitario_medio_por_region_pivot)

# Guardar el DataFrame en un archivo CSV en la carpeta actual
df.to_csv('data_sales.csv', index=False)

# Verificar que el archivo se haya creado correctamente
import os
print("\nArchivos en la carpeta actual:", os.listdir())
````

1.   Haz una tabla de frecuencia de 'yearOfStart' y visualízala por año.
- Ordenar por año.
- Dibuja un gráfico de líneas desde 1997 hasta 2014. <= Pista: plt.xlim()

**Respuesta:**

````python

# Archiv0 de ejemplo
data = {
    'yearOfStart': [1997, 1997, 2000, 2000, 2005, 2005, 2010, 2010, 2014, 2014],
    'CoffeeShop': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
}
df = pd.DataFrame(data)

# Guardar el DataFrame
df.to_csv('data_coffeeshop.csv', index=False)
df = pd.read_csv('data_coffeeshop.csv')

# 1. Tabla de frecuencia de 'yearOfStart' y visualización por año
tabla_frecuencia = df['yearOfStart'].value_counts().sort_index()
print("Tabla de frecuencia de 'yearOfStart':")
print(tabla_frecuencia)

# 2. Dibujar un gráfico de líneas desde 1997 hasta 2014
plt.figure(figsize=(10, 6))
plt.plot(tabla_frecuencia.index, tabla_frecuencia.values, marker='o', linestyle='-')
plt.title('Frecuencia de apertura de cafeterías por año')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.xlim(1997, 2014)
plt.xticks(range(1997, 2015), rotation=45)
plt.show()
````
