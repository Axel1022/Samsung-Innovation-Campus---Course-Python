```python
from datetime import datetime
from dateutil.parser import parse

lista_de_fechas = ["04/01/2024", "01-28-2024", "2024-03-16", "05/24/2024", "07/30/2025"]

fechaHoy = datetime.now()
anio = fechaHoy.year
mes = fechaHoy.month
dia = fechaHoy.day

formatoHoy = datetime.strptime(f"{dia}/{mes}/{anio}", '%d/%m/%Y')

for feha in lista_de_fechas:
  feha = parse(feha)
  fecha_formateada = datetime.strptime(f"{feha.day}/{feha.month}/{feha.year}", '%d/%m/%Y')
  diferencia = fecha_formateada - formatoHoy
  print("*-"*15)
  print(f"La diferencia en dias de {formatoHoy} hasta {fecha_formateada} es de : {diferencia.days} dias.")
  print(f"{diferencia.days} dias son {diferencia.days // 7}.{diferencia.days % 7} semanas.")
  print()


'''

from datetime import datetime

fecha1 = datetime.strptime("4/1/2024", '%d/%m/%Y')
fecha2 = datetime.strptime("1/2/2024", '%d/%m/%Y')

diferencia = fecha2 - fecha1

# Obtiene la diferencia en días
dias_diferencia = diferencia.days

print(f"Diferencia en días: {dias_diferencia}")

'''
```

# Actividad: Operaciones con Data Frames


## Instrucciones


1. Dado el siguiente diccionario construya un objeto DataFrame.
2. Filtre los resultados por año de publicación.
3. Obtenga algunas columnas basado en que contenga alguna palabra en el nombre de la publicacion.


```python

import pandas as pd

publicaciones_literarias = {
    1605: "Don Quijote de la Mancha - Miguel de Cervantes",
    1865: "Alicia en el país de las maravillas - Lewis Carroll",
    1949: "1984 - George Orwell",
    1954: "El Señor de los Anillos - J.R.R. Tolkien",
    1997: "Harry Potter y la piedra filosofal - J.K. Rowling",
    2003: "El Código Da Vinci - Dan Brown",
    2002: "Inventado para probar que el filtro funcione bien - Dan Brown"
}

data = list(publicaciones_literarias.items())

df = pd.DataFrame(data, columns=["Año", "Publicación - Autor"])

print("DataFrame original:")
print(df)
print()

print("Filtrado por año:")
df_filtrado_año = df[df["Año"] == 2002]
print(df_filtrado_año)
print()

print("Filtrado por palabra:")
df_filtrado_palabra = df[df["Publicación - Autor"].str.contains("Anillos")]
print(df_filtrado_palabra)
```
